```python
#Read CSV
dr1549 = pd.read_csv('../Project3_Team1/Data/deaths-risk-factor-15-49years.csv')
dr5069 = pd.read_csv('../Project3_Team1/Data/deaths-risk-factor-50-69years.csv')
```
